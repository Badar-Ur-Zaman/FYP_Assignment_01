## **Task : 1**

```python
import numpy as np
import matplotlib.pyplot as plt
Pt = 1
n = [3, 3.5, 4]
dh = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]
dv = 2
N = 10000
average_power = []
for i in n:
    power_dh = []
    for j in dh:
        d = np.sqrt(j**2 + dv**2)
        X = np.sqrt(0.5) * np.random.randn(N)
        Y = np.sqrt(0.5) * np.random.randn(N)
        g = np.abs(X + 1j * Y)
        g_square = g**2
        Pr = Pt * g_square * d**(-i)
        average_power_i = np.mean(Pr)
        power_dh.append(average_power_i)
    average_power.append(power_dh)
plt.figure(figsize=(6, 6))
for idx in range(len(n)):
    plt.plot(dh, average_power[idx], label=f'n = {n[idx]}')
plt.xlabel('dh')
plt.ylabel('Average Received Power')
plt.title('Average Received Power vs. Horizontal Distance for Different Path Loss Exponents')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
u=2
prob = []
all_rates = np.zeros((1000, 100))
N = -96
transmit_power = 1
n = 3.5  # Path loss exponent

for i in range(1000):
    user_x = np.random.uniform(0, 500, 100)
    user_y = np.random.uniform(0, 500, 100)
    bs_x, bs_y = 0, 0

    euclidean_distance = np.sqrt((user_x - bs_x)**2 + (user_y - bs_y)**2)
    path_loss = 1 / (euclidean_distance**n)

    X = np.sqrt(0.5) * np.random.randn(100)
    Y = np.sqrt(0.5) * np.random.randn(100)
    channel = np.abs(X + 1j * Y)

    received_power = transmit_power * path_loss * channel**2
    snr = received_power / (10 ** (N / 10))
    rate = np.log2(1 + snr)

    all_rates[i, :] = rate
    outage_count = np.sum(rate < u)
    outage_prob = outage_count / 100
    prob.append(outage_prob)

print("Average Outage Probability over 1000 runs:")
print(np.mean(prob))

```

```python
#9. Total Outage Porobability

# Per-user outage probability
prob_u_i = np.sum(all_rates < u, axis=0) / 1000
print("Per-user Outage Probabilities:")
print(prob_u_i)

# Total (mean) outage probability across all users
total_outage_probability = np.mean(prob_u_i)
print("Total Outage Probability:")
print(total_outage_probability)

```
